```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/fake_news.csv')
data.head()
```

```python
data.shape
```

```python
data.info
```

```python
data.isna().sum()
```

```python
data = data.drop(['id'], axis=1)
```

```python
data['content'] = data['author']+' '+data['title']+' '+data['text']
```

```python
data = data.drop(['author', 'title', 'text'], axis=1)
```

```python
data.head()
```

```python
data['content'] = data['content'].astype(str).apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
data['content'] = data['content'].str.replace('[^\w\s]','')
```

```python
from nltk.corpus import stopwords
stop  = stopwords.words('english')
data['content'] = data['content'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

```python
!pip install textblob
```

```python
from nltk.stem import WordNetLemmatizer
from textblob import Word
data['content'] = data['content'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
data['content'].head()
```

```python
X=data['content']
y=data['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# splitting into training and testing data

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=45, stratify=y)
```

```python
#validate the shape of train and test dataset

print (X_train.shape)

print (y_train.shape)

print (X_test.shape)

print (y_test.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vect = TfidfVectorizer(analyzer='word', token_pattern=r'\w{1,}', max_features=5000)

tfidf_vect.fit(data['content'])

# Use X_train and X_test directly instead of trying to access 'content'
xtrain_tfidf = tfidf_vect.transform(X_train)

xtest_tfidf = tfidf_vect.transform(X_test)
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn import metrics
pclf = PassiveAggressiveClassifier()
pclf.fit(xtrain_tfidf, y_train)
predictions = pclf.predict(xtest_tfidf)
print(metrics.classification_report(y_test, predictions))
```

```python
print(metrics.confusion_matrix(y_test,predictions))
```

```python
from sklearn.neural_network import MLPClassifier

mlpclf = MLPClassifier(hidden_layer_sizes=(256,64,16),

                       activation = 'relu',

                       solver = 'adam')

mlpclf.fit(xtrain_tfidf, y_train)

predictions = mlpclf.predict(xtest_tfidf)

print(metrics.classification_report(y_test, predictions))
```

```python
print(metrics.confusion_matrix(y_test,predictions))
```

```python
import pickle
pickle.dump(pclf, open('fakenews1.pkl', 'wb'))
```

```python
# load the model from disk
loaded_model = pickle.load(open("fakenews1.pkl", 'rb'))

# Transform X_test using the same TfidfVectorizer used during training
xtest_tfidf = tfidf_vect.transform(X_test)  # Assuming tfidf_vect is still available in your environment

# Calculate the score using the transformed data
result = loaded_model.score(xtest_tfidf, y_test)
print(result)
```
